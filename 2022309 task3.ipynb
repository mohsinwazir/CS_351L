```python
import matplotlib.pyplot as plt
import random

# Task details
tasks = {
    'Task 1': 2,  # Task duration in hours
    'Task 2': 1,
    'Task 3': 3,
    'Task 4': 2,
    'Task 5': 1,
    'Task 6': 4,
}

# Constraint: No two tasks can overlap
# Let's assume the workday is from 9 AM to 5 PM (8 hours)
workday_start = 9
workday_end = 17

# Schedule solution: time slots for each task
def schedule_tasks(tasks):
    scheduled_tasks = {}
    start_time = workday_start

    for task, duration in tasks.items():
        if start_time + duration <= workday_end:
            scheduled_tasks[task] = (start_time, start_time + duration)
            start_time += duration
        else:
            print(f"Cannot fit {task} in the workday!")
            break

    return scheduled_tasks

# Schedule the tasks
scheduled_tasks = schedule_tasks(tasks)

# Visualization of task scheduling using Matplotlib
def visualize_schedule(scheduled_tasks):
    fig, ax = plt.subplots(figsize=(10, 5))

    for i, (task, (start, end)) in enumerate(scheduled_tasks.items()):
        ax.barh(task, end - start, left=start, height=0.6, color=f'C{i+1}', edgecolor='black')

    ax.set_xlim(workday_start, workday_end)
    ax.set_xlabel('Time (Hours)')
    ax.set_ylabel('Tasks')
    ax.set_title('Task Scheduling Timeline')

    # Set x-ticks for time
    ax.set_xticks(range(workday_start, workday_end+1))

    plt.grid(True)
    plt.show()

# Run the visualization
visualize_schedule(scheduled_tasks)
```
